```python
from preprocessing import Validator
validator = Validator()
```

```python
# 800 , 0.04 : 7.475 / 17.59
# 1600, 0.03 : 7.414 / 17.58
# 2000, 0.01 : 7.615 / 17.66

import gc
gc.collect()
params = {
    'objective': 'l2',
    'verbosity': -1,
    'n_iter': 5000,
    'lambda_l1': 0.8942112689465215,
    'lambda_l2': 6.4122663335284305,
    #'cat_l2': 7.696733748814076,
    'learning_rate': 0.01, #0.054754587005265434,
    'max_depth': 11,
    'num_leaves': 273,
    'colsample_bytree': 0.6,#0.5488343345823052,
    'colsample_bynode': 0.8,#0.7200134511260632,
    'min_data_in_leaf': 6,
    'max_cat_threshold': 952,
}
validator.build_model(params=params)
```

```python
import pandas as pd
columns = validator.valid.loc[:1].drop(['sales', 'date', 'weight'], axis=1).columns.to_list()
split = validator.model.feature_importance(importance_type='split')
gain = validator.model.feature_importance(importance_type='gain')
imps = pd.DataFrame({'feature': columns, 'gain': gain, 'split': split})
imps['gain_per_split'] = (imps['gain'] / imps['split']).fillna(0)
imps = imps.sort_values('gain_per_split', ascending=False).reset_index(drop=True)
display(imps.head(30))
```
